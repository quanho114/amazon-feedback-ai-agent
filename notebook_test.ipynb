```python
import pickle
import pandas as pd

# Load dataframe và embeddings
df = pd.read_csv("data/processed/reviews_clean.csv")
with open("data/embeddings/reviews_embeddings.pkl", "rb") as f:
    embeddings = pickle.load(f)

# Chọn review đầu tiên
print("Review text:", df['clean_review'].iloc[0])
print("Embedding vector:", embeddings[0])
print("Dimension:", len(embeddings[0]))
```

```python
import numpy as np

emb_array = np.array(embeddings)

print("Shape:", emb_array.shape)       # (số review, 384)
print("Min:", emb_array.min())
print("Max:", emb_array.max())
print("Mean:", emb_array.mean())
```
