```python
import os
import joblib
from sentence_transformers import SentenceTransformer

# Nếu chạy notebook, dùng cwd hoặc chỉ định trực tiếp
BASE_DIR = os.getcwd()  # hoặc "/home/ho-minh-quan/Downloads/Amazon-review-AI"

clf_path = os.path.join(BASE_DIR, "models/sentiment_clf.pkl")
embedder_path = os.path.join(BASE_DIR, "models/sentence_embedder")

clf = joblib.load(clf_path)
embedder = SentenceTransformer(embedder_path)

from utils.text_cleaning import clean_text

reviews = ["the web amazon is so bad, I feel it very slow!", "I love buy many product on Amazon shop"]
reviews_clean = [clean_text(r) for r in reviews]

X_new = embedder.encode(reviews_clean)
y_new = clf.predict(X_new)

for r, s in zip(reviews, y_new):
    print(f"Review: {r}\nPredicted sentiment: {s}\n")
```
